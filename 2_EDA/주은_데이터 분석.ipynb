# 단변량 분석

```python
def uni(col, data, bins = 20):
    plt.figure(figsize = (8,5))
    sns.histplot(x = col, data = data, bins = bins)
    plt.show()
```

### 구별 분석

```python
for i in list(df_gu):
    if i != '자치구':
        uni(i, df_gu)
```

### 행정동별 분석

```python
for i in list(df_dong):
    if i != '행정동':
        uni(i, df_dong)
```

# 이변량 분석

```python
# 구별
def mul(col):
    tmp = df_gu.groupby(['자치구'])[col].mean().to_frame().sort_values(by = col)
    
    plt.figure(figsize = (16,5))
    sns.barplot(x = tmp.index, y = col, data = tmp)
    plt.show()
    
# 행정동별    
def mul2(col):
    tmp = df_dong.groupby(["행정동"])[col].mean().to_frame().sort_values(by = col)
    
    plt.figure(figsize = (16,5))
    sns.barplot(x = tmp.index, y = col, data = tmp)
    plt.show()    
```

### 구별 분석

```python
for i in list(df_gu):
    if i != '자치구':
        mul(i)
```

### 행정동별 분석

```python
for i in list(df_dong):
    if i != '행정동':
        mul2(i)
```

### 상관계수 0.5이상, pv 0.05보다 작으면 대립가설 맞음

```python
import scipy.stats as spst
```

```python
def pv_gu(col):
    print('<단속건수', '-', f'{col}>')
    print('상관계수: ',spst.pearsonr(df_gu['단속건수'],df_gu[col])[0])
    print('p-value: ',spst.pearsonr(df_gu['단속건수'],df_gu[col])[1])
    print()
    
def pv_dong(col):
    print('<단속건수', '-', f'{col}>')
    print('상관계수: ',spst.pearsonr(df_dong['단속건수'],df_dong[col])[0])
    print('p-value: ',spst.pearsonr(df_dong['단속건수'],df_dong[col])[1])
    print()
```
