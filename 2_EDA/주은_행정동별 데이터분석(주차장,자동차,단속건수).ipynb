```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rc('font',family='AppleGothic')
sns.set(font='AppleGothic',
       rc={'axes.unicode_minus':False},
       style='darkgrid')
```

```python
def show(data,var,target):
    a = data.groupby([var])[target].sum().to_frame().sort_values(by=target)
    
    plt.figure(figsize=(20, 3))
    sns.barplot(x = a.index, y = target, data = a,  palette='Blues')
    plt.title(target)
```

### 이변량 분석

- p-value값이 유의미한 것만 남김

```python
import scipy.stats as spst
```

```python
def display(var,target,data):
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=var, y=target, data=data)
    sns.lmplot(x=var, y=target, data=data)
    
    plt.xlabel(var)
    plt.ylabel(target)
    plt.show()
```

```python
var = '주차장 면수'
target = '단속건수'

s = list(spst.pearsonr(df[var],df[target]))
print(var,'&',target)
print()
print('상관계수: ',s[0])
print('p-value: ',s[1])

display(var,target,df)
```
